# Aplikacja, w której wykorzystamy nasz model do prognozy cen w lokalizacjach, w których ich nie znamy

Zacznijmy od wczytania pliku z charakterstyką lokalizacji, dla których chcemy obliczyć ceny mieszkań

```python
import pandas as pd

nieznane = pd.read_csv("Boston_nieznane.csv")

nieznane.head()
```

Jak widać, nie mamy tu kolumny 'medv', zawierającej ceny. Obliczym je korzystając z naszego modelu, dostępnego na dysku pod nazwą "nasz_regresor.sav".  

Zaczniemy od wczytania tego modelu.

```python
import pickle
nasz_regresor = pickle.load(open('nasz_regresor.sav', 'rb'))
```

```python
type(nasz_regresor)
```

A teraz wykorzystamy go do zaprognozowania wartości dla każdego z elementów w nieznanym zbiorze.

```python
prognozy = nasz_regresor.predict(nieznane)
```

```python
type(prognozy)
```

```python
# jak widać, prognozy to ndarray. Zamieńmy je na dataframe
prognozy_df = pd.DataFrame(prognozy, columns = ['prognoza'])
```

```python
# ... i scalmy z cechami lokalizacji
dane_z_prognozami = pd.concat([nieznane, prognozy_df], axis = 1)
```

```python
dane_z_prognozami
```

Udało się: w kolumnie 'prognoza' mamy wynik działania naszego algorytmu!

# Poniżej znajdziesz komplet kodu w jednej części

```python
# zaczynamy od niezbędnych importów
import pandas as pd
import pickle

nieznane = pd.read_csv("Boston_nieznane.csv") # tu pobieramy plik, który zawiera nasze zagadki
nasz_regresor = pickle.load(open('nasz_regresor.sav', 'rb')) # wczytujemy model z pliku
prognozy = nasz_regresor.predict(nieznane) # aplikujemy go do naszych nieznanych

prognozy_df = pd.DataFrame(prognozy, columns = ['prognoza']) # zamieniamy ndarray na dataframe
dane_z_prognozami = pd.concat([nieznane, prognozy_df], axis = 1) # i dołączamy do nieznanych
```
